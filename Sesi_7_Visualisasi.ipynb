```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel("https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Canada.xlsx", 
                       sheet_name='Canada by Citizenship',
                      skiprows = range(20),
                      skipfooter=2)
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.head(2)
```

```python
df_can.rename(columns={
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region'
}, inplace=True)
```

```python
df_can.head(2)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.head(2)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980, 2014)))
```

## Histogram

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years].plot(kind='hist')
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head(3)
```

```python
df_t.plot(kind='hist',
         figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# modification
# get the x-tick values
count, bin_edges = np.histogram(df_t, 15)
# define colors
colors = ['coral', 'darkslateblue','mediumseagreen']
#adding buffer
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

df_t.plot(kind='hist',
         figsize=(10, 6),
         bins = 15,
         alpha=0.6,
         xticks=bin_edges,
          xlim=(xmin, xmax),
          stacked=True,
         color=colors)

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

## Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5, 6),
                           autopct='%1.1f%%')
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral',
              'lightskyblue', 'lightgreen', 'plum']
explode_list = [0.1, 0, 0, 0, 0.1, 0.2]

df_continents['Total'].plot(kind='pie',
                           figsize=(15, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                           labels=None, #turn off labels
                            pctdistance=1.17,
                            colors=colors_list,
                            explode = explode_list
                           )

plt.title('Immigration to Canada by Continent', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Subplots

```python
fig = plt.figure()

ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

df_CI.plot(kind='box',
          figsize=(15, 6),
          vert=False,
          ax=ax0)
ax0.set_title('Box plot of Immigrants from India and China')
ax0.set_ylabel('Number of Immigrants')
ax0.set_xlabel('Countries')

df_CI.plot(kind='line',
          figsize=(15, 6),
          ax=ax1)
ax1.set_title('Line plot of Immigrants from India and China')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Countries')
```

## Scatter Plots

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.head()
```

```python
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.head()
```

```python
df_tot.columns = ['years', 'total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter',
           x='years',
           y='total',
           figsize=(10, 6))

plt.title('Total Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
```
