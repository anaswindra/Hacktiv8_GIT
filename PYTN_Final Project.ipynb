## Mental Health in Tech Industry Introduction
Currently over 1400 responses, the ongoing 2016 survey aims to measure attitudes towards mental health in the tech workplace, and examine the frequency of mental health disorders among tech workers.
This mental health is viewed within the tech/IT workplace, and the prevalence of certain mental health disorders within the tech industry. The Open Sourcing Mental Illness team of volunteers will use this data to drive our work in raising awareness and improving conditions for those with mental health disorders in the IT workplace.


**Attribute Information:**

There are 63 attributte

1. self_emp : Are you self-employed? 
2. emp_size : How many employees does your company or organization have? 
3. tech_comp : Is your employer primarily a tech company/organization?
4. tech_role : Is your primary role within your company related to tech/IT? 
5. mh_benefits_coverage : Does your employer provide mental health benefits as part of healthcare coverage?
6. mh_coverage_awareness : Do you know the options for mental health care available under your employer-provided coverage? 
7. mh_emp_dscs : Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?
8. seek_help : Does your employer offer resources to learn more about mental health concerns and options for seeking help?
9. anonymity : Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources provided by your employer?
10. medical_leave : If a mental health issue prompted you to request a medical leave from work, asking for that leave would be:
11. mh_neg_conseq_comp : Do you think that discussing a mental health disorder with your employer would have negative consequences?
12. ph_neg_conseq_comp : Do you think that discussing a physical health issue with your employer would have negative consequences?
13. mh_comfort_cowork_dscs : Would you feel comfortable discussing a mental health disorder with your coworkers?
14. mh_comfort_supervisor_dscs : Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?
15. comp_mh_vs_ph : Do you feel that your employer takes mental health as seriously as physical health?
16. mh_cowork_open_neg_conseq : Have you heard of or observed negative consequences for co-workers who have been open about mental health issues in your workplace?

18. mh_res_seek_help : Do you know local or online resources to seek help for a mental health disorder?

20. diag_rev_neg_impact : If you have revealed a mental health issue to a client or business contact, do you believe this has impacted you negatively?
21. diag_rev_cowork : If you have been diagnosed or treated for a mental health disorder, do you ever reveal this to coworkers or employees?
22. diag_rev_neg_impact_cowork : If you have revealed a mental health issue to a coworker or employee, do you believe this has impacted you negatively?
23. mh_prod_impact : Do you believe your productivity is ever affected by a mental health issue?
24. mh_prod_impact_perc : If yes, what percentage of your work time (time performing primary or secondary job functions) is affected by a mental health issue?
25. prev_employers : Do you have previous employers?
26. prev_mh_benefits : Have your previous employers provided mental health benefits?
27. prev_benefit_awareness : Were you aware of the options for mental health care provided by your previous employers?
28. prev_comp_dscs : Did your previous employers ever formally discuss mental health (as part of a wellness campaign or other official communication)?
29. prev_resource_seek_help : Did your previous employers provide resources to learn more about mental health issues and how to seek help?
30. prev_anonymity : Was your anonymity protected if you chose to take advantage of mental health or substance abuse treatment resources with previous employers?
31. prev_mh_neg_conseq_dscs : Do you think that discussing a mental health disorder with previous employers would have negative consequences?
32. prev_ph_neg_conseq_dscs : Do you think that discussing a physical health issue with previous employers would have negative consequences?
33. prev_mh_cowork_dscs : Would you have been willing to discuss a mental health issue with your previous co-workers?
34. prev_mh_supervisor_dscs : Would you have been willing to discuss a mental health issue with your direct supervisor(s)?
35. prev_mh_vs_ph : Did you feel that your previous employers took mental health as seriously as physical health?
36. prev_neg_cowork_conseq : Did you hear of or observe negative consequences for co-workers with mental health issues in your previous workplaces?
37. ph_interview : Would you be willing to bring up a physical health issue with a potential employer in an interview?
38. ph_why : Why or why not?
39. mh_interview : Would you bring up a mental health issue with a potential employer in an interview?
40. mh_why : Why or why not?.1
41. mh_hurt_career : Do you feel that being identified as a person with a mental health issue would hurt your career?
42. mh_suffered_neg_conseq : Do you think that team members/co-workers would view you more negatively if they knew you suffered from a mental health issue?
43. ff_mh_sharing : How willing would you be to share with friends and family that you have a mental illness?
44. mh_bad_response_workplace : Have you observed or experienced an unsupportive or badly handled response to a mental health issue in your current or previous workplace?
45. mh_sharing_reveal_workplace : Have your observations of how another individual who discussed a mental health disorder made you less likely to reveal a mental health issue yourself in your current workplace?
46. mh_fam_history : Do you have a family history of mental illness?
47. mh_disorder_past : Have you had a mental health disorder in the past?
48. mh_disorder_currently : Do you currently have a mental health disorder?
49. yes_diag_cond : If yes, what condition(s) have you been diagnosed with?
50. maybe_diag_cond : If maybe, what condition(s) do you believe you have?
51. mh_diag_prof : Have you been diagnosed with a mental health condition by a medical professional?
52. mh_diag_prof_cond : If so, what condition(s) were you diagnosed with?
53. mh_treatment_prof : Have you ever sought treatment for a mental health issue from a mental health professional?
54. mh_treat_impact : If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?
55. mh_not_treat_impact : If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?
56. age : What is your age?
57. gender : What is your gender?
58. country_live : What country do you live in?
59. US_territory_live : What US state or territory do you live in?
60. country_work : What country do you work in?

62. work_position : Which of the following best describes your work position?
63. work_remotely : Do you work remotely?

### Classification

```python
# drop y feature from x dataset, so it's just contain attribute values
x = df_emp6.drop('mh_diag_prof', 1)

# y variable contain y feature
y = df_emp6.mh_diag_prof
```

```python
x = pd.get_dummies(x)
x.head(3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,
                                                   y,
                                                   test_size=0.3,
                                                   random_state=13)
```

**Logistic Regression**

```python
# import Logistic Regression module
from sklearn.linear_model import LogisticRegression

# create logistic regression model using liblinear solver
LR = LogisticRegression(solver='liblinear')
# set the training data
LR.fit(x_train, y_train)
# get the prediction
pred_LR = LR.predict(x_test)

# get the accuracy and confussion matrix
acc_LR = accuracy_score(y_test, pred_LR)
print(acc_LR)
print(confusion_matrix(y_test, pred_LR))
print(classification_report(y_test, pred_LR))
```

**KNN**

```python
# import K-Nearest Neighbors module
from sklearn.neighbors import KNeighborsClassifier

# create the KNN model and predict the testing
KNN = KNeighborsClassifier()
KNN.fit(x_train, y_train)
pred_KNN = KNN.predict(x_test)

# get the accuracy and confussion matrix
acc_KNN = accuracy_score(y_test, pred_KNN)
print(acc_KNN)
print(confusion_matrix(y_test, pred_KNN))
```

**SVM**

```python
# import Support Vector Machine module
from sklearn import svm

# create the SVM model and predict the testing
modelSVM = svm.SVC()
modelSVM.fit(x_train, y_train)
pred_SVM = modelSVM.predict(x_test)

# get the accuracy and confussion matrix
acc_SVM = accuracy_score(y_test, pred_SVM)
print(acc_SVM)
print(confusion_matrix(y_test, pred_SVM))
```

**Decision Tree**

```python
# import Decision Tree module
from sklearn.tree import DecisionTreeClassifier

# create the Decision Tree model and predict the testing
DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)
pred_DT = DT.predict(x_test)

# get the accuracy and confussion matrix
acc_DT = accuracy_score(y_test, pred_DT)
print(acc_DT)
print(confusion_matrix(y_test, pred_DT))
```

**Random Forest**

```python
# import Random Forest module
from sklearn.ensemble import RandomForestClassifier

# create the Random Forest model and predict the testing
RF = RandomForestClassifier()
RF.fit(x_train, y_train)
pred_RF = RF.predict(x_test)

# get the accuracy and confussion matrix
acc_RF = accuracy_score(y_test, pred_RF)
print(acc_RF)
print(confusion_matrix(y_test, pred_RF))
```

**Naive Bayes**

```python
# import Naive Bayes module
from sklearn.naive_bayes import GaussianNB

# create the Naive Bayes model and predict the testing
NB = GaussianNB()
NB.fit(x_train, y_train)
pred_NB = NB.predict(x_test)

# get the accuracy and confussion matrix
acc_NB = accuracy_score(y_test, pred_NB)
print(acc_NB)
print(confusion_matrix(y_test, pred_NB))
```
