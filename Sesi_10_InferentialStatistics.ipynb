## Confidence Interval

range within which you expect the population parameter to be

```python
# semakin besar % maka rangenya juga semakin luas

import pandas as pd
import numpy as np

df = pd.read_csv('dataset/heart.csv')
```

```python
df.head()
```

target column to detect if there are heart attack or not

```python
df['Sex1'] = df.sex.replace({1: 'Male', 0:'Female'})
```

```python
df.head()
```

```python
df.groupby('Sex1').agg({'chol': [np.mean, np.std, np.size]})
```

```python
mean_fe = 261.45 #mean cholesterol of female
sd = 64.4 # standard deviation for female population
n = 312 # total number of female
z = 1.96 # 95% confidence, z-score from z table
```

```python
# calculate the standard error
se = sd/np.sqrt(n)
```

```python
# construct the confidence intervals
low_CI = mean_fe - z * se
hi_CI = mean_fe + z * se

low_CI, hi_CI
```

Dengan confidence level sebanyak 95% percaya bahwa rata-rata tingkat kolesterol wanita ada di antara 254.303 sampai 268.596

### Another Example of CI

```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import numpy as np
import random
import warnings
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(rc={'figure.figsize': (13, 7.5)})
sns.set_context('talk')
```

```python
np.random.normal(loc=181, scale=24, size=6000)
```

```python
# generate data of people's mass in pounds using normal distribution

np.random.seed(42)
normal_dist_us_male_mass_pounds = np.random.normal(loc=181, scale=24, size=6000)
normal_dist_us_female_mass_pounds = np.random.normal(loc=132, scale=22, size=6500)
all_mass_values = np.concatenate((normal_dist_us_male_mass_pounds, normal_dist_us_female_mass_pounds), axis=0)
df_ppl_mass = pd.DataFrame(data={'us_ppl_mass_pounds': all_mass_values})

df_ppl_mass
```

```python
df.groupby('Sex1').agg({'chol': [np.mean, np.std, np.size]})
```

```python
df.loc[132]
```

```python
sns.distplot(df_ppl_mass['us_ppl_mass_pounds'], color='darkslategrey')
plt.xlabel('mass [pounds]', labelpad=14)
plt.ylabel('probability of occurence', labelpad=14)
plt.title('Distribution of Mass of People in US', y=1.015, fontsize=20);
```

This is not normal distribution because there's no symmetry in centre values

```python
# calculation population mean
pop_mean_mass = df_ppl_mass['us_ppl_mass_pounds'].mean()
pop_mean_mass
```

```python
# calculate population standard deviation
pop_std_dev_mass = df_ppl_mass['us_ppl_mass_pounds'].std()
pop_std_dev_mass
```

```python
# get 300 sample with random values
sample_means = []
n = 25
for sample in range(0, 300):
    # random sampling done with replacement
    sample_values = np.random.choice(a=df_ppl_mass['us_ppl_mass_pounds'], size=n)
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
# view distribution of sample means (otherwise called sampling distribution)
sns.distplot(sample_means)
plt.title("Distribution of sample means ($n=25$) of people's mass in pounds", y=1.015, fontsize=20)
plt.xlabel('sample mean mass [pounds]', labelpad=14)
plt.ylabel('frequency of occurence', labelpad=14)
```

Sample distribution looks normal and the curve looks symmetric than before. Median and mean almost has same values. Let's look the summary below.

```python
# calculate median of sample means
median_of_sample_means = np.median(sample_means)
median_of_sample_means
```

```python
# calculate mean of sample means
mean_of_sample_means = np.mean(sample_means)
mean_of_sample_means
```

```python
pop_mean_mass
```

```python
# calculate standard deviation of sample means
std_dev_of_sample_means = np.std(sample_means)
std_dev_of_sample_means
```

#### Equation for Standard Deviation of Sampling Distribution

Standard deviation from sample means = standard error (SE)

```python
standard_error = pop_std_dev_mass/np.sqrt(n)
standard_error
```

#### Critical Z-Scores

```python
tail = 1 - stats.norm.cdf(1.96) # z-score 95%
tail
```

```python
# the area under the curve between the z-critical scores is approximately 0.95
1-(tail *2)
```

```python
# calculate the boundaries
lower_95_bound = mean_of_sample_means - 1.96*standard_error
lower_95_bound
```

```python
upper_95_bound = mean_of_sample_means + 1.96*standard_error
upper_95_bound
```

```python
kde = stats.gaussian_kde(sample_means)
pos = np.linspace(np.min(sample_means), np.max(sample_means), 10000)
plt.plot(pos, kde(pos), color='teal')
shade = np.linspace(lower_95_bound, upper_95_bound, 300)
plt.fill_between(shade, kde(shade), alpha = 0.45, color='teal')
plt.text(x=154, y=.01, horizontalalignment='center', fontsize=17, 
         s="95% of sample means fall within\n this shaded area of plus or minus \n 1.96 z-scores from the mean",
        bbox=dict(facecolor='whitesmoke', boxstyle='round, pad=0.1'))
plt.title("Distribution of sample means ($n=25$) of people's mass in pounds", y=1.015, fontsize=20)
plt.xlabel('sample mean mass [pounds]', labelpad=14)
plt.ylabel('frequency of occurence', labelpad=14)
```

### Population Means
Q : apakah pria memiliki Body Mass Index rata-rata lebih tinggi secara signifikan daripada wanita ?

```python
da = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv')
```

```python
da.head()
```

```python
females = da[
    da['RIAGENDR'] == 2
]

males  = da[
    da['RIAGENDR'] == 1
]
```

```python
n1 = len(females)
mu1 = females['BMXBMI'].mean()
sd1= females['BMXBMI'].std()

n1, mu1, sd1
```

```python
n2 = len(males)
mu2 = males['BMXBMI'].mean()
sd2= males['BMXBMI'].std()

n2, mu2, sd2
```

```python
import statsmodels.api as sm
```

```python
sm.stats.ztest(
    females['BMXBMI'].dropna(),
    males['BMXBMI'].dropna()
)
```

p-value = 6.591, lebih kecil daripada alpha yaitu 0.05 atau 5%.  
Reject null hypothesis, bahwa BMI male sama dengan BMI female. menerima alternative hypothesis yaitu BMI male tidak sama dengan BMI female.

```python
# let's create the visualizations

plt.figure(figsize=(7, 4))
plt.title('Female BMI histogram', fontsize=16)
plt.hist(females['BMXBMI'].dropna(), edgecolor='k', color='pink', bins=25)
plt.show()

plt.figure(figsize=(7, 4))
plt.title('Male BMI histogram', fontsize=16)
plt.hist(males['BMXBMI'].dropna(), edgecolor='k', color='blue', bins=25)
plt.show()
```
